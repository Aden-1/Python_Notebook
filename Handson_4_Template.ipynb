## 6.1 Prime Number Checker
Create a program that checks whether a number is a prime number and displays its factors if it is not a prime number.

### Console:
```powershell
Prime Number Checker

Please enter an integer between 1 and 5000: 5
5 is a prime number.

Try again? (y/n): y

Please enter an integer between 1 and 5000: 6
6 is NOT a prime number.
It has 4 factors: 1 2 3 6

Try again? (y/n): y

Please enter an integer between 1 and 5000: 200
200 is NOT a prime number.
It has 12 factors: 1 2 4 5 8 10 20 25 40 50 100 200

Try again? (y/n): n

Bye!
```

### Specifications:
- A prime number is divisible by two factors (1 and itself). For example, 7 is a prime number because it is only divisible by 1 and 7.
- If the user enters an integer that's not between 1 and 5000, the program should display an error message.
- If the number is a prime number, the program should display a message.
If the number is not a prime number, the program should display a message. Then, it should display the number of factors for the number and a list of those factors.
- Store the factors for each number in a list.
- Use functions to organize the code for this program.


````python
def main():
    print("Prime Number Checker")
    while True:
        userInput = int(input("Please enter an integer between 1 and 5000"))
        uIn = validateInt(userInput)
        isPrime, factors = checkPrime(uIn)

        if(isPrime):
            print(uIn, "is a prime number.")
        else:
            print(uIn, "is NOT a prime number.")
            print("it has", len(factors), "factors.")
            print(factors)

        repeat = input("Try again? (y/n)")
        if(repeat.lower() == 'n'):
            print("Bye!")
            break

def validateInt(uIn):
    while((uIn <= 1) or (uIn > 5000)):
        print("Invalid integer. Please try again.")
        uIn = int(input("Please enter an integer between 1 and 5000"))

    return int(uIn)

def checkPrime(candidate):
    factors = []
    c = 0
    prime = True

    for i in range(1, candidate+1):
        if(candidate % i == 0):
            factors.append(i)
            c += 1

    if(c>2):
        prime = False

    if(prime):
        return prime, [1, candidate]
    else:
        return prime, factors


if __name__ == "__main__":
    main()
````

### Specifications:
- Use a list of lists to store the data for the contacts. Provide starting data for two or more contacts.
- For the `view` and `del` commands, display an error message if the user enters and invalid contact number.
- When you exit the program, all changes that you made to the contact list are lost.




````python
def main():
    contacts = [["Eric Idle", "eric@ericidle.com", "724-002-4232"],["John Smith","john@jsmith.co.uk","900-234-4343"]]
    while True:
        printMenu()
        userInput = input("Command: ")
        if(userInput.lower() == "list"):
            print("List of contacts")
            listContacts(contacts)
        elif(userInput.lower() == "view"):
            print("View a contact")
            viewContact(contacts)
        elif(userInput.lower() == "add"):
            print("Add a contact")
            addContact(contacts)
        elif(userInput.lower() == "del"):
            print("Delete a contact")
            deleteContact(contacts)
        elif(userInput.lower() == "exit"):
            print("Bye!")
            break
        else:
            print("Not a valid command. Please try again.")

def deleteContact(contacts):
    cNum = int(input("Enter number: "))
    while cNum > len(contacts) - 1 or cNum < 0:
        print("Error that number is not in the contacts list")
        cNum = int(input("Enter number: "))
    print(contacts[cNum][0], "was deleted")
    contacts.pop(cNum)

def addContact(contacts):
    name = input("Name: ")
    email = input("Email: ")
    phone = input("Phone: ")
    newContact = [name, email, phone]
    contacts.append(newContact)
    print(name, "was added.")

def viewContact(contacts):
    cNum = int(input("Enter number: "))
    while cNum > len(contacts) - 1 or cNum < 0:
        print("Error that number is not in the contacts list")
        cNum = int(input("Enter number: "))
    print("Number:", cNum)
    print("Name:", contacts[cNum][0])
    print("Email:", contacts[cNum][1])
    print("Phone:", contacts[cNum][2])

def listContacts(contacts):
    contactCount = 0
    for contact in contacts:
            print(contactCount,". ", contact[0], sep="")
            contactCount += 1

def printMenu():
        print("COMMAND MENU")
        print("list - Display all contacts")
        print("view - View a contact")
        print("add - Add a contact")
        print("del - Delete a contact")
        print("exit - Exit program")

if __name__ == "__main__":
    main()
````
